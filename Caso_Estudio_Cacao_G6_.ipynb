# **Secado de cacao - G6**

##Datos

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Caso Base y Estudio Paramétrico

```python
T_aire=35
P_aire=1
PM_aire=29
PM_agua=18
R=0.082
H_final_c=0.04
H_inicial_c=0.6
m_c=350
Dab=2.6e-06
m_aire=1.895e-05
v_aire=10
L=3
A=2
Re_t=2e05
```

```python
H_rel_aire=np.linspace(0.7,0.8)#Variar la humedad relativa del aire 
tiempo_sec=[]
for Hrel in H_rel_aire:
  rho_aire=(P_aire*PM_aire)/(R*(T_aire+273))
  visc_aire=m_aire/rho_aire
  Re_L=(v_aire*L)/visc_aire
  Sc=visc_aire/Dab
  kc_prom=(0.664*Dab*(Re_t**(1/2))*(Sc**(1/3))+0.0365*Dab*(Sc**(1/3))*((Re_L**(4/5))-(Re_t**(4/5))))/L
  Psat_i=(Hrel/100)*(6.11*np.log((17.27*T_aire)/(237.3+T_aire)))
  Cas=Psat_i/(R*(T_aire+273))
  Pa=Hrel*Psat_i
  Ca=Pa/(R*(T_aire+273))
  Na_prom=kc_prom*(Cas-Ca)
  Area=L*A
  Wa_prom=Na_prom*Area
  m_agua_i=H_inicial_c*100
  m_cahum_i=m_agua_i+100
  m_w_i=(m_agua_i/m_cahum_i)*m_c
  mfinal_agua=H_final_c*100
  mfinal_cachum=mfinal_agua+100
  m_aguafinal=(mfinal_agua/mfinal_cachum)*m_c
  m_aguaevap=m_w_i-m_aguafinal
  mol_aguaevap=m_aguaevap/PM_agua
  tiempo_seci=(mol_aguaevap/Wa_prom)/(24*3600)
  tiempo_sec.append(tiempo_seci)
tiempo_secado=np.array(tiempo_sec)
plt.plot(H_rel_aire,tiempo_secado)
plt.xlabel('Humedad relativa del aire (%)')
plt.ylabel('Tiempo secado (días)')
plt.title('Estudio paramétrico: Humedad relativa')
plt.grid()
plt.show()
```
